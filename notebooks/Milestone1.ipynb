```python
def view_game(season_year, type_game, game_id, event_id):
    """
    Print the current widget value in short sentence
    """
    year_str = str(season_year)
    t_g_str = str(type_game).zfill(2)
    if t_g_str == '02':
        try:
            df = data_2017_2020.copy()
            df2 = df.filter(regex=(year_str + t_g_str))
            g_id = year_str + t_g_str + str(game_id).zfill(4)
            game_d = df2[g_id]['gameData']
            live_d = df2[g_id]['liveData']
            t_scores = live_d['plays']['allPlays']
            score = pd.DataFrame(columns = ['Home','Away'],index=['Teams','Goals'])
            score['Home']['Teams']=game_d['teams']['home']['abbreviation']
            score['Away']['Teams']=game_d['teams']['away']['abbreviation']
            score['Home']['Goals']=t_scores[(event_id-1)]['about']['goals']['home']
            score['Away']['Goals']=t_scores[(event_id-1)]['about']['goals']['away']
            coordinates = t_scores[(event_id-1)]['coordinates']
            des_event = t_scores[(event_id-1)]['result']['event']
            h_team = game_d['teams']['home']['name']
            a_team = game_d['teams']['away']['name']
            if coordinates != {}:
                team_event = t_scores[(event_id-1)]['team']['name']
                x = coordinates['x']
                y = coordinates['y']
                if x < 0:
                    if team_event == h_team:
                        right = h_team
                        left = a_team
                    else:
                        right = a_team
                        left = h_team
                else:
                    if team_event == h_team:
                        left = h_team
                        right = a_team
                    else:
                        left = a_team
                        right = h_team

                plot_event(coordinates, left, right)
            print(f'Current Season Game id seen is:{g_id}')
            print(game_d['datetime']['dateTime'])
            print('Game ID: ' + str(game_id) + "; " +  str(game_d['teams']['home']['abbreviation']) + " (home) vs " + str(game_d['teams']['away']['abbreviation']) + " (away)")
            print(score)
            print(f'Current viewed game: {game_id}')
            print(f'Current viewed event: {event_id}')
            print(des_event)
            if coordinates != {}:
                print(team_event)
        except:
            print("NO GAME DATA")
    else:
        try:
            df = data_2017_2020.copy()
            df2 = df.filter(regex=(year_str + t_g_str))
            po_id = list(df2.columns)[(game_id-1)]
            game_d = df2[po_id]['gameData']
            live_d = df2[po_id]['liveData']
            if live_d['plays']['allPlays'] != []:
                t_scores = live_d['plays']['allPlays']
                score = pd.DataFrame(columns = ['Home','Away'],index=['Teams','Goals'])
                score['Home']['Teams']=game_d['teams']['home']['abbreviation']
                score['Away']['Teams']=game_d['teams']['away']['abbreviation']
                score['Home']['Goals']=t_scores[(event_id-1)]['about']['goals']['home']
                score['Away']['Goals']=t_scores[(event_id-1)]['about']['goals']['away']
                coordinates = t_scores[(event_id-1)]['coordinates']
                des_event = t_scores[(event_id-1)]['result']['event']
                h_team = game_d['teams']['home']['name']
                a_team = game_d['teams']['away']['name']
                if coordinates != {}:
                    team_event = t_scores[(event_id-1)]['team']['name']
                    x = coordinates['x']
                    y = coordinates['y']
                    if x < 0:
                        if team_event == h_team:
                            right = h_team
                            left = a_team
                        else:
                            right = a_team
                            left = h_team
                    else:
                        if team_event == h_team:
                            left = h_team
                            right = a_team
                        else:
                            left = a_team
                            right = h_team

                    plot_event(coordinates, left, right)
                print(f'Current Playoff Game id seen:{po_id}')
                print(game_d['datetime']['dateTime'])
                print('Game ID: ' + str(game_id) + "; " +  str(game_d['teams']['home']['abbreviation']) + " (home) vs " + str(game_d['teams']['away']['abbreviation']) + " (away)")
                print(score)
                print(f'Current viewed game: {game_id}')
                print(f'Current viewed event: {event_id}')
                print(des_event)
                if coordinates != {}:
                    print(team_event)
            else:
                print("It was scheduled but there were no plays recorded")
        except:
            print("NO GAME DATA")
```
