```python
a = [-75,0,30,25]
b = [30,10,30,10]
c = ['shot' , 'pause' , 'goal' , 'faul']
i_d = [1 , 2 , 3 , 4]
```

```python
def varf(e,r,t,p):
    
    for x,y,t,u in zip(e,r,t,p):
        img = plt.imread('nhl_rink.png')
        fig, ax = plt.subplots(figsize=(12, 6))
        plt.xlim(-100, 100)
        plt.ylim(-45, 45)
        ax.imshow(img, extent=[-100, 100, -45, 45], aspect='auto')
        plt.plot(x, y, marker="o", markersize=20, markeredgecolor="red", markerfacecolor="green")
        plt.title(t)
        plt.show()
```
